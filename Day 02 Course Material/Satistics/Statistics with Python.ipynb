# Mode

```python
data=[12,52,52,68,78,47,96,22,36,65,44,9]
dataa=pd.Series([12,52,52,68,78,47,96,22,36,65,44,9])
print('Mode: ',dataa.mode())
```

# Percentile

```python
# First Quartile
print('Q1: ',dataa.quantile(0.25))
```

```python
# Second Quartile
print('Q2: ',dataa.quantile(0.5))
```

```python
# Third Quartile
print('Q3: ',dataa.quantile(0.75))
```

# IQR Value

```python
q1=dataa.quantile(0.25)
q3=dataa.quantile(0.75)
```

```python
IQR=q3-q1
IQR
```

```python
Max_range=q3+1.5*IQR
Min_range=q1-1.5*IQR
```

```python
Min_range
```

```python
Max_range
```
